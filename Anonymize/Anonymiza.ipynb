```python
import pandas as pd
import numpy as np
from datetime import timedelta
```

```python
df = pd.read_csv("dataset.csv", delimiter=';')
```

```python
df.head()
```

```python
df.dtypes
```

```python
#SUPRESSÃO

df.drop(columns='CPF')
df['Nome']=df['Nome'].apply(lambda x: x.split(" ")[0])
df.head()
```

```python
#MASKING ou PII Masking

# df['CPF'].apply(lambda x: x.replace(' \d\d:\d\d', '', regex=True))
df['CPF'] = df['CPF'].str.replace('\d\d\d', 'XXX', regex=True)
```

```python
df.head()
```

```python
#Generalização
df['Data Nascimento'] = pd.to_datetime(df['Data Nascimento'],format='%d/%m/%Y')

df['Data Nascimento'] = pd.DatetimeIndex(df['Data Nascimento']).year
df.head()
```

```python
#Perturbation

df['Data Nascimento'] = pd. to_datetime(df['Data Nascimento'],format='%d/%m/%Y')

df['Data Nascimento'] = pd. DatetimeIndex(df['Data Nascimento'])+timedelta(days=31)
df.head()
```

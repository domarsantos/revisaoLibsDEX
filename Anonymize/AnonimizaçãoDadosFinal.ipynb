```python
import pandas as pd
from datetime import timedelta
from faker import Faker
```

```python
df = pd.read_csv("dataset.csv", delimiter=';')
```

```python
df.head()
```

```python
#SUPRESSÃO

df['Nome'] = df['Nome'].apply(lambda x: x.split(" ")[0])
```

```python
df.head()
```

```python
#MASKING ou PII Mask

df['CPF'] = df['CPF'].replace('\d\d\d', 'XXX', regex=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
# GENERALIZAÇÃO

df['Data Nascimento'] = pd.to_datetime(df['Data Nascimento'], format='%d/%m/%Y')
df['Data Nascimento'] = pd.DatetimeIndex(df['Data Nascimento']).year
df.head()
```

```python
# PERTURBATION

df['Data Nascimento'] = pd.to_datetime(df['Data Nascimento'], format='%d/%m/%Y')
df['Data Nascimento'] = pd.DatetimeIndex(df['Data Nascimento']) + timedelta(days=12)
```

```python
df.head()
```

```python
# Dado Sintetico ou Fake

fake = Faker('pt-BR')
df['E-mail'] = df['E-mail'].apply(lambda x: fake.email())
df.head()
```

```python
#HASH

df['CEP'] = df['CEP'].apply(lambda x: hash(x))
df.head()
```

```python
df[['Salary', 'Salary Unit']] = df['Salary'].str.split('/', expand=True)
```

```python
df['Salary Unit'].unique()
```

```python
df['Salary'] =  df['Salary'].apply(lambda x: x.lstrip('₹').replace(',', ''))
```

```python
df['Salary'] = df['Salary'].astype(int)
```
